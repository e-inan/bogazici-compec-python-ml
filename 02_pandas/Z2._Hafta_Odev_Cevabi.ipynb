# Satışlar ve Ürünleri birleştirme

# Inner Join two dataframes

```python
order_products_df = pd.merge(order_items_df, products_df, left_on = "orderItemProductId", right_on = "productId", how = "inner")
```

# Sipariş_Ürün birleşimini kategori ile birleştirelim

```python
order_product_cat_df = pd.merge(order_products_df, categories_df, left_on = "productCategoryId", right_on = "categoryId", how = "inner")
```

# Kategori isimlerine göre en çok satış tutarına sahip ilk 20 ürün

```python
order_product_cat_df.groupby(['categoryName']).orderItemSubTotal.agg(['sum']).sort_values("sum", ascending=False).iloc[:20,:].head(30)
```

# Sonuçlardan yeni bir df oluşturma

```python
df_grp = order_product_cat_df.groupby(["categoryName"]).orderItemSubTotal \
.agg(['sum']) \
.sort_values("sum", ascending=False) \
.iloc[:20,:]
```

# Grafik

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = df_grp.index
y = df_grp['sum']


y_pos = range(len(x))
plt.figure(figsize=(16,8)) 
plt.bar(y_pos, y, align='center', alpha=0.5) # alfa saydamlık derecesi
plt.xticks(y_pos, x)
plt.xlabel('Kategoriler')
plt.xticks(rotation=45)
plt.ylabel('Satış Tutarı')
plt.title('Kategorilere Göre Toplam Satışlar (İlk 20)')
 
plt.show()
```

# Cevap-2

```python
# orders tablosu içinde sipariş durum bilgisi var. İptal mi, bekliyor mu vs.
orders_df = pd.read_csv("D:\\OneDrive\\egitim\\verilen\\bogazici-compec-python-ml\\Datasets\\retail_db\\orders.csv")
```

```python
order_items_orders_df = pd.merge(order_items_df, orders_df, left_on = "orderItemOrderId", right_on = "orderId", how = "inner")
```

```python
order_items_orders_df.head()
```

```python
order_items_orders_products_df = pd.merge(order_items_orders_df, products_df, left_on = "orderItemProductId", right_on = "productId", how = "inner")
```

```python
order_items_orders_products_df.head()
```

```python
order_items_orders_products_categories_df = pd.merge(order_items_orders_products_df, categories_df, 
                                                     left_on="productCategoryId", right_on="categoryId", how="inner")
```

```python
order_items_orders_products_categories_df.head()
```

```python
grouped_df = order_items_orders_products_categories_df[order_items_orders_df['orderStatus'] == 'COMPLETE'] \
.groupby(['categoryName']).orderItemSubTotal \
.agg(['sum']).sort_values('sum', ascending=False).iloc[:20,:]
```

```python
grouped_df.head(50)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = grouped_df.index
y = grouped_df['sum']


y_pos = range(len(x))
plt.figure(figsize=(16,8)) 
plt.bar(y_pos, y, align='center', alpha=0.5) # alfa saydamlık derecesi
plt.xticks(y_pos, x)
plt.xlabel('Kategoriler')
plt.xticks(rotation=45)
plt.ylabel('Satış Tutarı')
plt.title('Kategorilere Göre Toplam Satışlar (İlk 20)')
 
plt.show()
```
